```python

import random
import pandas as pd
from datetime import datetime
from datetime import timedelta
import numpy as np
from random import choice, randint
random.seed(42)

customers = list(range(10000,99999))
sales_reps = ['Rep_A', 'Rep_B', 'Rep_C', 'Rep_D']
stages = ['Prospecting', 'Qualification', 'Proposal', 'Negotiation', 'Closed Won', 'Closed Lost']
lead_sources = ['Web', 'Referral', 'Social Media', 'Direct Mail', 'Email Campaign']
lead_quality = ['High', 'Medium', 'Low']
close_dates = [datetime.today() - timedelta(days=randint(1, 365)) for _ in range(5)]

num_records = 100  

data = {
    'Customer_ID': [choice(customers) for _ in range(num_records)],
    'Sales_Rep': [choice(sales_reps) for _ in range(num_records)],
    'Deal_Value': [randint(1000, 50000) for _ in range(num_records)],
    'Stage': [choice(stages) for _ in range(num_records)],
    'Close_Date': [choice(close_dates) for _ in range(num_records)],
    'Lead_Source': [choice(lead_sources) for _ in range(num_records)],
    'Lead_Quality': [choice(lead_quality) for _ in range(num_records)],
    'Lead_Creation_Date': [datetime.today() - timedelta(days=randint(10, 100)) for _ in range(num_records)],
    'Deal_Probability': [round(np.random.uniform(0.2, 1.0), 2) for _ in range(num_records)],
    'Deal_Notes': ['Potential high-value deal' if np.random.rand() > 0.5 else 'Needs more nurturing' for _ in range(num_records)],
}


df_crm = pd.DataFrame(data)

df_crm.to_csv('crm_sales_data.csv', index=False)

print("First 5 records of the CRM Sales Dataset:")
print(df_crm.head())


```

```python
import pandas as pd
import numpy as np
from random import choice, randint
from datetime import datetime, timedelta

call_ids = list(range(50000,99999))
speakers = ['Rep_A', 'Rep_B', 'Rep_C', 'Rep_D', 'Customer01','Customer02']
sentiments = ['Positive', 'Neutral', 'Negative']
keywords = ['Pricing', 'Product Features', 'Support', 'Onboarding', 'Objections', 'Renewal', 'Discount', 'Demo', 'Integration', 'Competitor']
transcripts = [
    "I think the product is a great fit for our needs.",
    "Could you clarify how the integration works?",
    "I'm worried about the pricing, it's higher than I expected.",
    "Can we get a discount if we sign up for a longer term?",
    "How does the onboarding process look?",
    "I'm not sure about the customer support response time.",
    "The features look impressive, but I need more details on integration.",
    "Are there any case studies you can share with us?",
    "The product is good, but we're also looking at a competitor."
]


num_records = 50 

data = {
    'Call_ID': [choice(call_ids) for _ in range(num_records)],
    'Timestamp': [datetime.today() - timedelta(minutes=randint(0, 1440)) for _ in range(num_records)],  # Random timestamps within the last day
    'Speaker': [choice(speakers) for _ in range(num_records)],
    'Transcript': [choice(transcripts) for _ in range(num_records)],
    'Sentiment': [choice(sentiments) for _ in range(num_records)],
    'Keywords': [', '.join(np.random.choice(keywords, size=randint(1, 3), replace=False)) for _ in range(num_records)],
    'Duration': [randint(10, 180) for _ in range(num_records)],  # Random duration between 10s to 180s
}


df_gong_transcripts = pd.DataFrame(data)

df_gong_transcripts.to_csv('gong_transcript_data.csv', index=False)

print("First 5 records of the Gong Transcript Dataset:")
print(df_gong_transcripts.head())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

stage_counts = df_crm.groupby(['Sales_Rep', 'Stage']).size().reset_index(name='Count')

# Step 2: Pivot the data for easier plotting
pivot_stage = stage_counts.pivot(index='Sales_Rep', columns='Stage', values='Count').fillna(0)

# Step 3: Plot the results using a stacked bar chart
plt.figure(figsize=(10, 6))
pivot_stage.plot(kind='bar', stacked=True)

# Adding labels and title
plt.title('Comparison of Sales Reps by Stage', fontsize=16)
plt.xlabel('Sales Rep', fontsize=12)
plt.ylabel('Count of Deals in Each Stage', fontsize=12)
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.legend(title="Stages")
plt.tight_layout()
plt.show()
```
